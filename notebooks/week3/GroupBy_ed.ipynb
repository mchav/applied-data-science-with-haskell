# Grouping Data

Sometimes we want to select data based on groups and understand aggregated data on a group level. We have seen that even though Pandas allows us to iterate over every row in a dataframe, it is generally very slow to do so. Fortunately Pandas has a groupby() function to speed up such task. The idea behind the groupby() function is  that it takes some dataframe, splits it into chunks based on some key values, applies computation on those  chunks, then combines the results back together into another dataframe. In pandas this is referred to as the split-apply-combine pattern.

## Splitting

```haskell
import qualified DataFrame as D
```

```haskell
rawDf <- D.readCsv "./datasets/census.csv"

D.take 10 rawDf
```

```haskell
import qualified DataFrame.Functions as F
import DataFrame.Functions ((.==))
F.declareColumns rawDf

df = D.filterWhere (sumlev .== 50) rawDf

D.take 10 df
```

In the first example for groupby() I want to use the census date. Let's get a list of the unique states, then we can iterate over all the states and for each state we reduce the data frame and calculate the average.

Let's run such task for 3 times and time it. For this we'll use the cell magic function %%timeit

```haskell
import Control.Monad
import qualified Data.Set as S
import qualified Data.Text as T

import DataFrame ((|>))

states = S.fromList (D.columnAsList @T.Text (F.name stname) df)

forM_ [1..3] $ \_ -> do
    forM_ states $ \state -> do
        let avg = df |> D.filterWhere (stname .== F.lit state) |> D.mean (F.col "CENSUS2010POP")
        print $ "Counties in state " ++ T.unpack state ++ " have an average population of " ++ show avg
```

If you scroll down to the bottom of that output you can see it takes a fair bit of time to finish. Now let's try another approach using groupby

```haskell
forM_ [1..3] $ \_ -> do
    print (df |> D.groupBy [F.name stname]
              |> D.aggregate [F.mean census2010pop `F.as` "average_population"])
```

This is faster and much more readable.

## Applying

To this point we have applied very simple processing to our data after splitting, really just outputting some print statements to demonstrate how the splitting works. The pandas developers have three broad categories of data processing to happen during the apply step, Aggregation of group data, Transformation of group data, and Filtration of group data

### Aggregation

Let's continue our groupby exploration on AirBnb data.

```haskell
airbnb <- D.readCsv "./datasets/listings.csv"
```

```haskell
D.describeColumns airbnb
```

```haskell
F.declareColumns airbnb
```

```haskell
import DataFrame.Functions ((.=))

airbnb |> D.groupBy ["cancellation_policy"]
       |> D.aggregate ["review_scores_value" .= F.mean review_scores_value]
       |> D.take 5
```

That didn't seem to work at all. The compiler wouldn't let us run the code. The issue is actually in the function that we sent to aggregate. F.mean does not work on optional columns! We have a version that works on optional columns called meanMaybe.

```haskell
import DataFrame.Functions ((.=))

airbnb |> D.groupBy ["cancellation_policy"]
       |> D.aggregate ["review_scores_value" .= F.meanMaybe (F.col "review_scores_value")]
       |> D.sortBy [D.Desc "review_scores_value"]
```

We can just extend this dictionary to aggregate by multiple functions or multiple columns.

```haskell
airbnb |> D.groupBy ["cancellation_policy"]
       |> D.aggregate [ "review_scores_value" .= F.meanMaybe review_scores_value
                      , "reviews_per_month" .= F.stddevMaybe reviews_per_month
                      ]
```

Take a moment to make sure you understand the previous cell, since it's somewhat complex. First we're doing a group by on the dataframe by the column "cancellation_policy". This creates a new GroupedDataFrame. Then we are invoking the aggregate function on that frame. The aggregate function is going to apply one or more functions we specify to the grouped dataframe and return a single row per group.
