# `DataFrame` Manipulation

Now that you know the basics of what makes up a dataframe, lets look at how we might actually clean some messy data. Now, there are many different approaches you can take to clean data, so this lecture is just one example of how you might tackle a problem.

Haskell programmers will often suggest that there many ways the language can be used to solve a particular problem. But that some are more appropriate than others. The best solutions are celebrated as Idiomatic Python and there are lots of great examples of this on StackOverflow and other websites.

A sort of sub-language within Haskell, dataframe has its own set of idioms.

```haskell
import qualified DataFrame as D

df <- D.readCsv "datasets/census.csv"

D.take 10 df
```

```haskell
-- The first of these is called method chaining.
-- The general idea behind method chaining is that every method on an object 
-- returns a reference to that object. The beauty of this is that you can 
-- condense many different operations on a DataFrame, for instance, into one line 
-- or at least one statement of code.
-- Here's an example of two pieces of code in pandas using our census data.

import qualified DataFrame.Functions as F
import DataFrame.Functions ((.==))
import DataFrame ((|>))

df
    |> D.filterWhere (F.col @Int "SUMLEV" .== 50)
    |> D.rename "ESTIMATESBASE2010" "Estimates Base 2010"
    |> D.take 10
```
